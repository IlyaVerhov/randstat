Need to simulate a game against a lottery machine like "777". The player pays 1 dollar, after which a random integer uniformly distributed from 0 to 999 falls out. For some values of the number, the player receives a win.

WINS:

*   777: 200 dol.
*   999: 100 dol.
*   555: 50 dol.
*   333: 15 dol.
*   111: 10 dol.
*   *77: 5 dol.
*   **7: 3 dol.
*   *00: 2 dol.
*   **0: 1 dol.

```python
from random import randint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
WINS = {
    777: 200,
    999: 100,
    555: 50,
    333: 15,
    111: 10
}
```

```python
def check_tail(current):
    if current % 100 == 0:
        return 2
    elif current % 100 == 77:
        return 5
    elif current % 10 == 0:
        return 1
    elif current % 10 == 7:
        return 3
    return 0

def get_curr():
    return randint(0, 999)

def main():
    received = 0
    spent = 0
    attempts = 1000000
    MMC = 0
    while True:
        if spent <= attempts:
            print(current := get_curr())
            if current in WINS:
                received += WINS[current]
            elif tail := check_tail(current):
                received += tail
            spent += 1
            MMC = received / spent
            print('Денег выигранно: ', received)
            print('Денег проигранно: ', spent)
            print('MMC: ', round(MMC, 2)) # В целом ограничение в две цифры после запятой можно считать некорректным, но заметил я это после того, как запустил код выполняться на час.
            continue
        elif spent == attempts:
            break

main()
```

Выше смоделировано миллион игр, показатель ММК составил 0.8. Если пользоваться теоритическими инструментами и посчитать вероятность выйгрыша по формуле полной вероятности результат бы был таким же - вероятность победы составила бы порядка 20%.
